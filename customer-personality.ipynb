```python
# Data Analysis
import numpy as np
import pandas as pd
import random

# Data Visualization
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px
import plotly.figure_factory as ff
%matplotlib inline

# Data
data = pd.read_csv("../input/customer-personality-analysis/marketing_campaign.csv", sep='\t')
data = data.dropna()
```

<a id="section-two"></a>
# Question 1
**What is the average rate of responses per visits?**

```python
data['rate'] = data['Response'] / data['NumWebVisitsMonth']
average_rate = round(data['rate'].mean(), 2)
fig = px.histogram(data, x="rate", histnorm='percent', title='Ratio of Responses per Visits')
fig.show()
print('The average rate:', average_rate*100, '%')
```

So we can see that we got 4 responses from each 100 visits, but let's ask the question again in more accurate way what is the average ratio between the number of customer visits per their responses?

```python
cur_data = data.query('NumWebVisitsMonth > 0')
average_rate = round(cur_data.query('Response > 0').shape[0] / cur_data.shape[0], 2)
print('The average rate:', average_rate*100, '%')
```

Here we've got a more logical result because when we do so we exclude the number of repeated visits.

<a id="section-three"></a>
# Question 2
**Does income have a big effect on this rate or our products didn't attract the customer?**

```python
response = data.query('Response == 1')
non_response = data.query('Response == 0')
fig = go.Figure()
fig.add_trace(go.Violin(x=response['Income'], line_color='lightseagreen', name='Response', y0=0))
fig.add_trace(go.Violin(x=non_response['Income'], line_color='red', name= 'Don\'t response', y0=0))

fig.update_traces(orientation='h', side='positive', meanline_visible=False)
fig.update_layout(xaxis_showgrid=False, xaxis_zeroline=False)

fig.update_layout(title='<b>Income distribution (Response Vs Non-response)<b>',
                  xaxis_title='Income',
                  titlefont={'size': 18},
                  width=600,
                  height=400,
                  template="plotly_dark",
                  showlegend=True,
                  paper_bgcolor="lightgray",
                  plot_bgcolor='lightgray', 
                  font=dict(
                      color ='black',
                      )
                  )
fig.update_layout(xaxis_range=[0,150000])
fig.show()
```

So we can see that income doesn't have a huge effect on responding and this is a good news because if it has this will mean that the products is expensive because many customers don't buy because of the high prices, but this doesn't happen so we have a different reason.

# Question 3
**Is there a negative relationship between customers who buy sweets and who buy meets or fish (healthy food), so we can categorize each customer?**

For answering this questio we must see first what is the relashionship between fish products and meet products because if there is a week relationship among them we must see the relationship between each of them with sweet products.

```python
def scatter(data, var1, var2, title, xlabel, ylabel, alpha=1):
    plt.scatter(data[var1], data[var2], alpha=alpha)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.show()

scatter(data, 'MntMeatProducts', 'MntFishProducts', 'Relationship between Meat and Fish', 'Meat', 'Fish', alpha=0.1)
```

We can see that it's a positive relationship, but it's week so we will check each of them with sweet.

```python
scatter(data, 'MntMeatProducts', 'MntSweetProducts', 'Relationship between Meat and Sweet', 'Meat', 'Sweets', alpha=0.1)
```

**There is a strange thing here!**

We can see that the relationship between meat and sweets is as stronge as its with fish. If we find similar thing with fish we must avoid categorizing customers as healthy or unhealthy.

```python
scatter(data, 'MntFishProducts', 'MntSweetProducts', 'Relationship between Fish and Sweet', 'Fish', 'Sweets', alpha=0.1)
```

And .. yes, it's also strong positive relationship so we will avoid categorizing customers

Let's see them all at once.

```python
data_cur = data[['MntMeatProducts', 'MntFishProducts', 'MntSweetProducts']]
corr = data_cur.corr()
fig = ff.create_annotated_heatmap(
    z=corr.to_numpy().round(2),
    x=list(corr.index.values),
    y=list(corr.columns.values),       
    xgap=3, ygap=3,
    zmin=-1, zmax=1,
    colorscale='earth',
    colorbar_thickness=30,
    colorbar_ticklen=3,
)
fig.update_layout(title_text='<b>Correlation Matrix (cont. features)<b>',
                  title_x=0.5,
                  titlefont={'size': 24},
                  width=550, height=550,
                  xaxis_showgrid=False,
                  xaxis={'side': 'bottom'},
                  yaxis_showgrid=False,
                  yaxis_autorange='reversed',                   
                  paper_bgcolor=None,
                  )
fig.show()
```

# Question 4
**Is there a strong positive relationship between number of kids and sweet products do he have?**

```python
data['Kidhome'] = data['Kidhome'].astype('str')
fig = go.Figure()
fig.add_trace(go.Scatter(y=data['MntSweetProducts'], x=data['Kidhome'], mode='markers', name="#Kids", marker_color='lightseagreen'))
fig.update_xaxes(title_text="Number of kids")
fig.update_yaxes(title_text="Sweet Products")
fig.update_layout(title_text="<b>Relation between Sweets and number of kids")
fig.show()
```

**Another strange insight here!**

There is a negative relationship between sweet products and the number of kids customer has.

<a id="section-six"></a>
# Question 5
**Assume people are either children (0-12), teens (13-20), youth (21-35) or old (36+) what is the most age range buy from us?**

```python
data['Dt_Customer'] = pd.to_datetime(data['Dt_Customer'])
data['Age'] = data['Dt_Customer'].dt.year - data['Year_Birth']
data['Age_Range'] = 'Child'
data.loc[data['Age'] >= 13, 'Age_Range'] = 'Teens'
data.loc[data['Age'] >= 21, 'Age_Range'] = "Youth"
data.loc[data['Age'] >= 36, 'Age_Range'] = "Old"

fig = go.Figure()
fig.add_trace(
    go.Pie(
        labels=data['Age_Range'],
        values=None,
        hole=.4,
        title='Age Range',
        titlefont={'color':None, 'size': 18},
        )
    )
fig.update_traces(
    hoverinfo='label+value',
    textinfo='label+percent',
    textfont_size=12,
    marker=dict(
        colors=['lightgray', 'lightseagreen', 'lightblue'],
        line=dict(color='#000000',
                  width=2)
        )
    )
fig.show()
```

So our age range of interest is 36+
